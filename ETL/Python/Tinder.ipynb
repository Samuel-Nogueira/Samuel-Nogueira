Nesta primeira parte é onde é feito o Login na conta do Tinder.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager


email = 'Seu e-mail'
senha = 'Sua SENHA'
caminho_save = 'arquivo o caminho onde o arquivo será salvo.'
site = "https://tinder.com/"

options = webdriver.ChromeOptions()
options.add_argument("--headless")

driver = webdriver.Chrome(ChromeDriverManager().install(), chrome_options=options)
driver.get(site)

time.sleep(2)
driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div/main/div[1]/div/div/div/div/header/div/div[2]/div[2]").click()

time.sleep(2)
main_page = driver.current_window_handle 

time.sleep(2)
driver.find_element_by_xpath('/html/body/div[2]/div/div/div[1]/div/div[3]/span/div[2]/button/span[2]').click() 
for handle in driver.window_handles: 
    if handle != main_page: 
        login_page = handle 

time.sleep(2)

driver.switch_to.window(login_page) 
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/form/div/div[1]/div/input').send_keys(email)
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/form/div/div[2]/div/input').send_keys(senha)
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/form/div/div[3]/label[2]/input').click()

time.sleep(2)

driver.switch_to.window(main_page) 

time.sleep(5)

driver.find_element_by_xpath('/html/body/div[2]/div/div/div/div/div[3]/button[1]').click()
driver.find_element_by_xpath('/html/body/div[1]/div/div[2]/div/div/div[1]/button/span').click()

time.sleep(5)

driver.find_element_by_xpath('/html/body/div[2]/div/div/div/div/div[3]/button[1]').click()
```

Nesta segunda o código irá interagir com a pagina, coletandos os dados e realizando o Macth nos usuários que aparecem. 

```python
lista = []
for x in range(10):
    print(x)
    try:
        time.sleep(2)
        driver.find_element_by_xpath('//*[@id="o-1335420887"]/div/div/div[2]/button[2]').click()
        print('Pulou o popup 1')
        
    except Exception:        
        try:
            time.sleep(2)
            driver.find_element_by_xpath('//*[@id="o-1335420887"]/div/div/button[2]').click()
            print('Pulou o popup 2')
            
        except Exception:
            try:
                time.sleep(3)
                driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/main/div[1]/div/div/div[1]/div/div/div[3]/div[3]/button').click()
                print('Abriu o Perfil')
                
                time.sleep(3)
                print('Iniciou a coleta dos dados')
                info = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div/main/div[1]/div/div/div[1]/div[1]/div/div[2]')

                nome = info.find_element(By.CSS_SELECTOR, "h1").text
                idade = info.find_element(By.CSS_SELECTOR, "span").text

                try:
                    lista_cadastro = []
                    for x in range(1,10):
                            path = 'div/div/div[2]/div['+ str(x) +']'
                            cadastro  = info.find_element(By.XPATH, path).text
                            lista_cadastro.append(cadastro)

                except Exception:

                    bio = []
                    bio.append(info.find_element(By.XPATH, 'div[2]').text)

                    tupla = (nome,idade,lista_cadastro,bio)
                    dicionario = {'nome' : tupla[0],
                    'idade' : tupla[1],
                    'Cadastro' : tupla[2],
                    'bio' : tupla[3]
                    }

                    lista.append(dicionario)
                    print('Terminou a coleta dos dados.')
                    
                    time.sleep(3)
                    driver.find_element_by_xpath('//*[@id="o-1556761323"]/div/div[1]/div/main/div[1]/div/div/div[1]/div[2]/div/div/div[4]/button').click()
                    print('Curtiu o Perfil')
                
            except Exception:
                driver.refresh()
                print('Precisou atualizar')

tabela = pd.DataFrame(lista)
tabela.to_csv(r'C:\Users\lsnog\Documents\Python\Scrap\teste.csv')

print('Concluido!')
```
